```python


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('C:\\Users\\Q-Admin\\Documents\\pokemon.csv',index_col=0)
```

```python
df.head()
```

```python
sns.lmplot(x='Attack',y='Defense',data=df,fit_reg=False,hue='Generation')
plt.xlim(0,None)
plt.ylim(0,None)
```

```python
sns.boxplot(data=df)
```

```python
stats_df=df.drop(['Total','Generation','Legendary'],axis=1)
```

```python
sns.boxplot(data=stats_df)
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(15,12))
sns.violinplot(x='Type 1',y='Attack',data=df)
pkmn_type_colors = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                   ]
```

```python
plt.figure(figsize=(15,12))
sns.violinplot(x='Type 1',y='Attack',palette=pkmn_type_colors,data=df)
sns.violinplot(x='Type 2',y='Defense',palette=pkmn_type_colors,data=df)
```

```python
plt.figure(figsize=(15,12))
sns.swarmplot(x='Type 1',y='Attack',palette=pkmn_type_colors,data=df)
```

```python
plt.figure(figsize=(15,12))
sns.violinplot(x='Type 1',y='Attack',inner=None,palette=pkmn_type_colors,data=df)
sns.swarmplot(x='Type 1',y='Attack',color='k',alpha=0.7,data=df)
plt.title('Attack By Type')
```

```python
melted_df=pd.melt(stats_df,id_vars=['Name','Type 1','Type 2'],var_name='Stats')
melted_df.head()
```

```python
plt.figure(figsize=(15,12))
sns.swarmplot(x='Stats',y='value',data=melted_df,hue='Type 1',split=True,palette=pkmn_type_colors)
plt.legend(bbox_to_anchor=(1,1),loc=2)
```

```python
corr=stats_df.corr()
sns.heatmap(corr)
```

```python
sns.distplot(df.Attack)
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Type 1',data=df,palette=pkmn_type_colors)
plt.xticks(rotation=-45)
```
